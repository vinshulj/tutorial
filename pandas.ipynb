```python
class demo:
    def __init__(self):
        self.x = 10   # calls __setattr__

    #Called for EVERY attribute access
    def __getattribute__(self, name):
        print(f"__getattribute__ called for: {name}")
        return object.__getattribute__(self, name)

    #Called ONLY if attribute NOT found
    def __getattr__(self, name):
        print(f"__getattr__ called for missing: {name}")
        return "default"

    #Called on attribute assignment
    def __setattr__(self, name, value):
        print(f"__setattr__ called: {name} = {value}")
        object.__setattr__(self, name, value)

    # Called on attribute deletion
    def __delattr__(self, name):
        print(f"__delattr__ called for: {name}")
        object.__delattr__(self, name)
```

```python
class MyList:
    def __init__(self):
        self.data = {}

    # obj[key]
    def __getitem__(self, key):
        print(f"__getitem__ called with key: {key}")
        return self.data[key]

    # obj[key] = value
    def __setitem__(self, key, value):
        print(f"__setitem__ called: {key} = {value}")
        self.data[key] = value

    # del obj[key]
    def __delitem__(self, key):
        print(f"__delitem__ called for key: {key}")
        del self.data[key]

    # item in obj
    def __contains__(self, item):
        print(f"__contains__ called for item: {item}")
        return item in self.data.values()
```
